outliers

```python
# Assuming numerical columns  is your DataFrame
columns_to_change = ["ApplicantIncome","CoapplicantIncome","LoanAmount","Loan_Amount_Term","Credit_History"]

# Create box plots for each column to visualize outliers
for column in columns_to_change:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=loandata[column])
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
q1=np.percentile(loandata['LoanAmount'],25,interpolation="midpoint")
q2=np.percentile(loandata['LoanAmount'],50,interpolation="midpoint")
q3=np.percentile(loandata['LoanAmount'],75,interpolation="midpoint")

q1
```

```python
q2
```

```python
q3
```

```python
IQR=q3-q1
IQR
```

```python
min_val=q1 - 1.5*IQR
max_val=q3 + 1.5*IQR
print(min_val,max_val)
```

```python
outliers=[]
for x in loandata['LoanAmount']:
  if ((x>max_val) or (x<min_val)):
    outliers.append(x)

outliers
```

```python
loandata['LoanAmount']=loandata['LoanAmount'].clip(min_val,max_val)
```
