```python

# Assuming housedata is your DataFrame
columns_to_change = ['Sale Price','No of Bedrooms',"No of Bathrooms","Flat Area (in Sqft)","Lot Area (in Sqft)","No of Floors","Overall Grade",
                     "Area of the House from Basement (in Sqft)","Basement Area (in Sqft)","Age of House (in Years)","Renovated Year","Zipcode","Latitude",
                     "Living Area after Renovation (in Sqft)","Lot Area after Renovation (in Sqft)"]

# Create box plots for each column to visualize outliers
for column in columns_to_change:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=housedata[column])
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python

q1=np.percentile(housedata['Sale Price'],25,interpolation="midpoint")
q2=np.percentile(housedata['Sale Price'],50,interpolation="midpoint")
q3=np.percentile(housedata['Sale Price'],75,interpolation="midpoint")

q1
```

```python
q2
```

```python
q3
```

```python

IQR=q3-q1
IQR
```

```python
min_val=q1 - 1.5*IQR
max_val=q3 + 1.5*IQR
print(min_val,max_val)
```

```python
outliers=[]
for x in housedata['Sale Price']:
  if ((x>max_val) or (x<min_val)):
    outliers.append(x)

outliers
```

```python
housedata['Sale Price']=housedata['Sale Price'].clip(min_val,max_val)
```
