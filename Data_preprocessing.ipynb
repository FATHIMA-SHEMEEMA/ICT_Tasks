# **#outliers**

```python
columns_to_change=['beer_servings','spirit_servings',
                   'wine_servings','total_litres_of_pure_alcohol']
```

```python
for col in columns_to_change:
  plt.figure()
  plt.boxplot(new[col])
```

```python
q1=np.percentile(new['spirit_servings'],25,interpolation="midpoint")
q2=np.percentile(new['spirit_servings'],50,interpolation="midpoint")
q3=np.percentile(new['spirit_servings'],75,interpolation="midpoint")

q1
```

```python
q2
```

```python
q3
```

```python
IQR=q3-q1
IQR
```

```python
min_val=q1 - 1.5*IQR
max_val=q3 + 1.5*IQR
print(min_val,max_val)
```

```python
outliers=[]
for x in new['spirit_servings']:
  if ((x>max_val) or (x<min_val)):
    outliers.append(x)

outliers
```

```python
new['spirit_servings']=new['spirit_servings'].clip(min_val,max_val)
```
